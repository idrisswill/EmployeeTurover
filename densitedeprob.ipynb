```python
# Densit de probabilit de numvotes des 2 classes d'employee
sns.distplot(data[data.stillExists==True]['numVotes'], hist=False, label='numvotes_true')
sns.distplot(data[data.stillExists==False]['numVotes'], hist=False, label='numvotes_false')
plt.legend()
plt.title('numvotes employee True and false')
```

```python
mean_numvote_true = np.mean(data[data.stillExists==True]['numVotes'])
mean_numvote_false = np.mean(data[data.stillExists==False]['numVotes'])
std_numvote_true = np.std(data[data.stillExists==True]['numVotes'])
std_numvote_false = np.std(data[data.stillExists==False]['numVotes'])
samples_norm_true = np.random.normal(mean_numvote_true,std_numvote_true, size = 100000)
samples_norm_false = np.random.normal(mean_numvote_false,std_numvote_false, size = 100000)
```
